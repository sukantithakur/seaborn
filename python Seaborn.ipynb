```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
```

```python
sns.relplot(x = "passengers", y = "month",data = a)
```

```python
sns.relplot(x = "passengers", y = "month",hue = "year", data = a)
plt.show()
```

```python
tips_data = sns.load_dataset("tips")
```

```python
sns.relplot(x= 'time',y = "tip", data = tips_data, kind = "line")
```

```python
#Categorical scatterplots
sns.catplot(x = "day",y = "total_bill",data = tips_data) # default kind = "strip"
```

```python
#Categorical distribution plots
sns.catplot(x = "day",y = "total_bill",data = tips_data,kind = "violin")
```

```python
#Categorical distribution plots
sns.catplot(x = "day",y = "total_bill",data = tips_data,kind = "boxen")
```

```python
#Categorical distribution plots
sns.catplot(x = "day",y = "total_bill",data = tips_data,kind = "box")
```

```python
#Categorical scatterplots
sns.catplot(x = "day",y = "total_bill",data = tips_data,kind = "swarm")
```

```python
#seaborn FacetGrid

iris_data = sns.load_dataset("iris")
iris_map = sns.FacetGrid(iris_data,col ="species")
iris_map.map(plt.hist,"sepal_length")
plt.show()
```

```python
#seaborn PairGrid
data = sns.load_dataset("flights")
flight_data = sns.PairGrid(data)
flight_data.map(plt.scatter)
plt.show()
```
